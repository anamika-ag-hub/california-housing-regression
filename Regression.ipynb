```python
#Loading and Preprocessing

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)

df['MedHouseVal'] = housing.target
df.head()
```

```python
df.isnull().sum()
```

```python
#Feature Scaling

X = df.drop('MedHouseVal', axis=1)
y = df['MedHouseVal']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

X_scaled.head()
```

The California Housing dataset was first loaded using sklearn and converted into a pandas DataFrame for easier handling and analysis.
The target variable (median house value) was added to the dataset as a separate column to clearly distinguish between input features and the output variable.
The dataset was checked for missing values using isnull(), and no missing values were found, so no cleaning was required.
Feature scaling was then applied using StandardScaler to normalize the input features. This ensures that all features are on the same scale, which helps improve the performance and accuracy of machine learning models, especially those sensitive to feature magnitude.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
#Linear Regression

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

Linear Regression finds a linear relationship between input features and output.
It is simple and works well for basic patterns.

```python
#Decision Tree Regressor

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
```

Decision Tree splits the data into branches based on conditions.
It can capture complex relationships but may overfit.

```python
#Model Evaluation and Comparison

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
def evaluate(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, mae, r2

# Evaluate models
lr_metrics = evaluate(y_test, y_pred_lr)
dt_metrics = evaluate(y_test, y_pred_dt)

print("Linear Regression:", lr_metrics)
print("Decision Tree:", dt_metrics)
```

```python
results = pd.DataFrame({
    "Model": ["Linear Regression", "Decision Tree"],
    "MSE": [lr_metrics[0], dt_metrics[0]],
    "MAE": [lr_metrics[1], dt_metrics[1]],
    "R2 Score": [lr_metrics[2], dt_metrics[2]]
})

results
```

The performance of Linear Regression and Decision Tree was compared using MSE, MAE, and R² score.

Decision Tree has lower MSE and MAE values compared to Linear Regression, which means it has less prediction error.

It also has a higher R² score, indicating that it explains the data better.

Therefore, Decision Tree performs better than Linear Regression for this dataset.
